## Bar Chart
A bar chart is used to depict the distribution of a categorical variable. In a bar chart, each level of the categorical variable is depicted with a bar, whose height indicates the frequency of data points that take on that level. A basic bar chart of frequencies can be created through the use of seaborn's countplot function:

## Pie Charts
A pie chart is a common univariate plot type that is used to depict relative frequencies for levels of a categorical variable. Frequencies in a pie chart are depicted as wedges drawn on a circle: the larger the angle or area, the more common the categorical value taken.

```python
diamonds['clarity'].value_counts()
```

```python
# code for the pie chart generation_id above bar chart
sorted_counts = diamonds['clarity'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False);
plt.axis('square');
```

## Histograms
A histogram is used to plot the distribution of a numeric variable. It's the quantitative version of the bar chart. However, rather than plot one bar for each unique numeric value, values are grouped into continuous bins, and one bar for each bin is plotted depicting the number. For instance, using the default settings for matplotlib's hist function:


```python
diamonds['price'].head(10)
```

```python
plt.hist(data=diamonds, x='price');
#matplotlib by default make 10 bins of values
```

```python
bins=np.arange(0, diamonds['price'].max()+5, 5) # start, end, and step size
plt.hist(data=diamonds, x='price', bins=bins);
```

```python
bins=np.arange(0, diamonds['depth'].max()+0.5, 0.5) # start, end, and step size
plt.hist(data=diamonds, x='depth', bins=bins);
```

#### Scales and Transformations
Certain data distributions will find themselves amenable to scale transformations. The most common example of this is data that follows an approximately log-normal distribution. This is data that, in their natural units, can look highly skewed: lots of points with low values, with a very long tail of data points with large values. However, after applying a logarithmic transform to the data, the data will follow a normal distribution.

```python
bins=np.arange(0, diamonds['table'].max()+40, 40) # start, end, and step size
plt.hist(data=diamonds, x='table', bins=bins);
```

```python
bins=np.arange(0, diamonds['table'].max()+40, 40) # start, end, and step size
plt.hist(data=diamonds, x='table', bins=bins);
plt.xscale('log');
```

## Scatterplots
If we want to inspect the relationship between two numeric variables, the standard choice of plot is the scatterplot. In a scatterplot, each data point is plotted individually as a point, its x-position corresponding to one feature value and its y-position corresponding to the second. One basic way of creating a scatterplot is through Matplotlib's scatter function:

```python
plt.scatter(data=diamonds, x='carat',y='price')
plt.xlabel('Carat')
plt.ylabel('Price')
```

```python
sns.regplot(data=diamonds, x='carat',y='price')
plt.xlabel('Carat')
plt.ylabel('Price')
```

#### Overplotting, Transparency, and Jitter
If we have a very large number of points to plot or our numeric variables are discrete-valued, then it is possible that using a scatterplot straightforwardly will not be informative. The visualization will suffer from overplotting, where the high amount of overlap in points makes it difficult to see the actual relationship between the plotted variables.

```python
#add jitter
sns.regplot(data=diamonds, x='table',y='price',x_jitter=0.3);
```

```python
#Add Transparency
sns.regplot(data=diamonds, x='depth',y='price',x_jitter=0.3,
           scatter_kws={'alpha': 1/20});

#darker area shows the overlap of the data ; this is area Where the main bulk of the data lies.
```

## Heat Maps
A heat map is a 2-d version of the histogram that can be used as an alternative to a scatterplot. Like a scatterplot, the values of the two numeric variables to be plotted are placed on the plot axes. Similar to a histogram, the plotting area is divided into a grid and the number of points in each grid rectangle is added up. Since there won't be room for bar heights, counts are indicated instead by grid cell color. A heat map can be implemented with Matplotlib's hist2d function.

```python
plt.hist2d(data=diamonds, x='x',y='y')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.hist2d(data=fuel_econ, x='carat',y='price',cmin=0.5)
plt.colorbar()
plt.xlabel('Carat')
plt.ylabel('Price')
```

```python
plt.hist2d(data=diamonds, x='carat',y='price',cmin=0.5,cmap='viridis_r')
plt.colorbar()
plt.xlabel('Carat')
plt.ylabel('Price')
```

## Clustered Bar Charts
To depict the relationship between two categorical variables, we can extend the univariate bar chart seen in the previous lesson into a clustered bar chart. Like a standard bar chart, we still want to depict the count of data points in each group, but each group is now a combination of labels on two variables. So we want to organize the bars into an order that makes the plot easy to interpret. In a clustered bar chart, bars are organized into clusters based on levels of the first variable, and then bars are ordered consistently across the second variable within each cluster. This is easiest to see with an example, using seaborn's countplot function. To take the plot from univariate to bivariate, we add the second variable to be plotted under the "hue" argument:

```python
diamonds['clarity'].head()
```

```python
ct_counts = diamonds.groupby(['cut','clarity']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts.pivot(index='cut',columns='clarity',values='count')
```

```python
sns.countplot(data=diamonds,x='cut',hue='clarity')
plt.xticks(rotation=15)
```

```python
bins = np.arange(12,58+2,2)
g = sns.FacetGrid(data=diamonds,col='cut',col_wrap=3)
g.map(plt.hist,'depth',bins=bins);
```

```python
base_color = sns.color_palette()[0]
sns.barplot(data=fuel_econ,x='cut',y='depth',color=base_color)
plt.xticks(rotation=15)
plt.ylabel('Depth')
```

## Line plot
The line plot is a fairly common plot type that is used to plot the trend of one numeric variable against values of a second variable. In contrast to a scatterplot, where all data points are plotted, in a line plot, only one point is plotted for every unique x-value or bin of x-values (like a histogram). If there are multiple observations in an x-bin, then the y-value of the point plotted in the line plot will be a summary statistic (like mean or median) of the data in the bin. The plotted points are connected with a line that emphasizes the sequential or connected nature of the x-values.

```python
sns.pointplot(data = fuel_econ, x = 'cut', y = 'price')
plt.ylabel('Price')
plt.xticks(rotation=15)
```
